```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,3*np.pi / 2,0.005)
x
```

```python
y1 = np.sin(np.pi*x)
```

```python
y3 = np.sin(3*np.pi*x)/3
```

```python
y5 = np.sin(5*np.pi*x)/5
```

```python
y7 = np.sin(7*np.pi*x)/7
```

```python
y9 = np.sin(9*np.pi*x)/9
```

```python
y = y1 + y3 + y5 + y7 + y9
```

```python
fig, ax=plt.subplots()
ax.plot(y)
ax.set(xlabel='x', ylabel='y', title='Fourier Series 1 to 9')
ax.grid()
```

```python
y11 = np.sin(11*np.pi*x)/11
y13 = np.sin(13*np.pi*x)/13
y15 = np.sin(15*np.pi*x)/15
```

```python
yy = y11 + y13 + y15
```

```python
fig, ax=plt.subplots()
ax.plot(yy)
ax.set(xlabel='x', ylabel='y', title='Fourier Series 11 to 15')
ax.grid()
```

We notice a great difference between these two graphs. The first seems to be near -1 and 1 everywhere, except when x=200, 400, 600, etc. when it jumps to the other end. The second graph seems to alternate between -0.2 and 0.2 and the addidtion of these three functions seems to illustrate similar groupings of waves. In both cases, we see a continual increase and decrease of points throughout the graphs and a type of reflection.
